```python
path1 = "/kaggle/input/playground-series-s3e18/"
path2 = "/kaggle/input/multi-label-classification-of-enzyme-substrates/"

sample = pd.read_csv(path1+"sample_submission.csv")
train = pd.read_csv(path1+"train.csv",index_col='id')
test = pd.read_csv(path1+"test.csv",index_col='id')
original = pd.read_csv(path2+"original.csv",index_col='id')

# Combine playground dataset and synthetic dataset generated
train1 = pd.concat([train,original]).reset_index(drop=True).drop_duplicates()

id_test = test.index # id column required for submission file
target1 = 'EC1'
target2 = 'EC2'
```

# Check for duplicated rows

```python
features = [f for f in test.columns if not (f.startswith('EC') or f=='id')]
features1 = features + ['EC1', 'EC2']

dup_train = train.duplicated()
print(f'Duplicated number of rows in train set (features columns + EC1~EC6 columns) = {dup_train.sum()}')
dup_train = train[features].duplicated()
print(f'Duplicated number of rows in train set (features columns only) = {dup_train.sum()}')
dup_train = train[features1].duplicated()
print(f'Duplicated number of rows in train set (features columns + EC1~EC2 columns) = {dup_train.sum()}')
print("\n")

dup_test = test.duplicated()
print(f'Duplicated number of rows in test set (features columns only) = {dup_test.sum()}')
print("\n")

dup_original = original.duplicated()
print(f'Duplicated number of rows in original set (features columns + EC1~EC6 columns) = {dup_original.sum()}')
dup_original = original[features].duplicated()
print(f'Duplicated number of rows in original set (features columns only) = {dup_original.sum()}')
dup_original = original[features1].duplicated()
print(f'Duplicated number of rows in original set (features columns + EC1~EC2 columns) = {dup_original.sum()}')
print("\n")

dup_train1 = train1.duplicated()
print(f'Duplicated number of rows in train1 set (features columns + EC1~EC6 columns) = {dup_train1.sum()}')
dup_train1 = train1[features].duplicated()
print(f'Duplicated number of rows in train1 set (features columns only)= {dup_train1.sum()}')
dup_train1 = train1[features1].duplicated()
print(f'Duplicated number of rows in train1 set (features columns + EC1~EC2 columns) = {dup_train1.sum()}')
```

```python
dup_train_test = pd.concat([train[features],test[features]],axis=0).duplicated()
print(f'Duplicated number of rows between train and test sets (features columns only) = {dup_train_test.sum()}')
dup_train1_test = pd.concat([train1[features],test[features]],axis=0).duplicated()
print(f'Duplicated number of rows between train1 and test sets (features columns only) = {dup_train1_test.sum()}')
dup_ori_test = pd.concat([original[features],test[features]],axis=0).duplicated()
print(f'Duplicated number of rows between original and test sets (features columns only) = {dup_ori_test.sum()}')
print("\n")

dup_train_ori = pd.concat([train,original],axis=0).duplicated()
print(f'Duplicated number of rows between train and original sets (features columns + EC1~EC6 columns) = {dup_train_ori.sum()}')
dup_train_ori = pd.concat([train[features],original[features]],axis=0).duplicated()
print(f'Duplicated number of rows between train and original sets (features columns only) = {dup_train_ori.sum()}')
dup_train_ori = pd.concat([train[features1],original[features1]],axis=0).duplicated()
print(f'Duplicated number of rows between train and original sets (features columns + EC1~EC2 columns) = {dup_train_ori.sum()}')
```
