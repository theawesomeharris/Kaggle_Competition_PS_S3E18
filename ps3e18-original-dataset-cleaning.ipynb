# Check for integrity of EC1_EC2_EC3_EC4_EC5_EC6 columns across 3 original files


```python
col_check = ['CIDs', 'EC1_EC2_EC3_EC4_EC5_EC6']

check = pd.merge(original_desc[col_check],original_ecfp[col_check], on="CIDs", how='inner')
check = pd.merge(check,original_fcfp[col_check], on="CIDs", how='inner')
check = check.rename(columns={'EC1_EC2_EC3_EC4_EC5_EC6_x':'EC1_EC2_EC3_EC4_EC5_EC6_desc','EC1_EC2_EC3_EC4_EC5_EC6_y':'EC1_EC2_EC3_EC4_EC5_EC6_ecfp','EC1_EC2_EC3_EC4_EC5_EC6':'EC1_EC2_EC3_EC4_EC5_EC6_fcfp'})
check['identical'] = check.apply(lambda row: row['EC1_EC2_EC3_EC4_EC5_EC6_desc'] == row['EC1_EC2_EC3_EC4_EC5_EC6_ecfp'] == row['EC1_EC2_EC3_EC4_EC5_EC6_fcfp'], axis=1)
display(check.head(3))
percentage_false = (1 - check['identical'].astype(int).mean()) * 100
print(f"Percentage of False values: {percentage_false.round(2)}%")
```

Seems like the original dataset has integrity issues on the target values (EC1~EC6) across the 3 files.

# Data Cleaning

As observed above, most columns in train.csv is in mixed.desc.csv and the EC1 to EC6 is in both mixed_ecfp.csv and mixed_fcfp.csv files. Hence the cleaning steps are as follows:
1. Merge the interested columns from each file into one file
2. Clean the column names

```python
# Tidy up original datasets with 3 dataframes into 1 dataframe
col_desc = ['CIDs', 'BertzCT', 'Chi1', 'Chi1n', 'Chi1v', 'Chi2n', 'Chi2v', 'Chi3v',
        'Chi4n', 'EState_VSA1', 'EState_VSA2', 'ExactMolWt', 'FpDensityMorgan1',
        'FpDensityMorgan2', 'FpDensityMorgan3', 'HallKierAlpha',
        'HeavyAtomMolWt', 'Kappa3', 'MaxAbsEStateIndex', 'MinEStateIndex',
        'NumHeteroatoms', 'PEOE_VSA10', 'PEOE_VSA14', 'PEOE_VSA6', 'PEOE_VSA7',
        'PEOE_VSA8', 'SMR_VSA10', 'SMR_VSA5', 'SlogP_VSA3', 'VSA_EState9',
        'fr_COO', 'fr_COO2']
col_desc1 = ['CIDs', 'BertzCT', 'Chi1', 'Chi1n', 'Chi1v', 'Chi2n', 'Chi2v', 'Chi3v',
        'Chi4n', 'EState_VSA1', 'EState_VSA2', 'ExactMolWt', 'FpDensityMorgan1',
        'FpDensityMorgan2', 'FpDensityMorgan3', 'HallKierAlpha',
        'HeavyAtomMolWt', 'Kappa3', 'MaxAbsEStateIndex', 'MinEStateIndex',
        'NumHeteroatoms', 'PEOE_VSA10', 'PEOE_VSA14', 'PEOE_VSA6', 'PEOE_VSA7',
        'PEOE_VSA8', 'SMR_VSA10', 'SMR_VSA5', 'SlogP_VSA3', 'VSA_EState9',
        'fr_COO', 'fr_COO2', 'EC1_EC2_EC3_EC4_EC5_EC6']
col_ecfp = ['CIDs', 'EC1_EC2_EC3_EC4_EC5_EC6']

# Merge the two files with selected columns
original = pd.merge(original_desc[col_desc], original_ecfp[col_ecfp], on="CIDs", how='inner')
original_1 = original_desc[col_desc1]

# Rename "CIDs" into "id"
original = original.rename(columns={'CIDs':'id'})
original_1 = original_1.rename(columns={'CIDs':'id'})

# Split and drop "EC1_EC2_EC3_EC4_EC5_EC6"
original[['EC1', 'EC2', 'EC3', 'EC4', 'EC5', 'EC6']] = \
    original['EC1_EC2_EC3_EC4_EC5_EC6'].str.split('_', expand=True).astype(int)
original.drop('EC1_EC2_EC3_EC4_EC5_EC6', axis=1, inplace=True)

original_1[['EC1', 'EC2', 'EC3', 'EC4', 'EC5', 'EC6']] = \
    original_1['EC1_EC2_EC3_EC4_EC5_EC6'].str.split('_', expand=True).astype(int)
original_1.drop('EC1_EC2_EC3_EC4_EC5_EC6', axis=1, inplace=True)

display(original.shape), display(original.columns)
display(original.head(3)), display(original.describe().applymap("{:.2f}".format))
display(original.info())
```

# Check for duplicates in rows

```python
test = pd.read_csv('/kaggle/input/playground-series-s3e18/test.csv')
features = [f for f in test.columns if not (f.startswith('EC') or f=='id')]
features1 = features + ['EC1', 'EC2']
```

```python
dup_original = original.duplicated()
print(f'Duplicated number of rows in original set (features columns + EC1~EC6 columns from _ecfp file) = {dup_original.sum()}')
dup_original = original[features].duplicated()
print(f'Duplicated number of rows in original set (features columns only) = {dup_original.sum()}')
dup_original = original[features1].duplicated()
print(f'Duplicated number of rows in original set (features columns + EC1~EC2 columns from _ecfp file) = {dup_original.sum()}')
print("\n")


dup_original_1 = original_1.duplicated()
print(f'Duplicated number of rows in original set (features columns + EC1~EC6 columns from _desc file) = {dup_original_1.sum()}')
dup_original_1 = original_1[features].duplicated()
print(f'Duplicated number of rows in original set (features columns only) = {dup_original_1.sum()}')
dup_original_1 = original_1[features1].duplicated()
print(f'Duplicated number of rows in original set (features columns + EC1~EC2 columns from _desc file) = {dup_original_1.sum()}')
print("\n")
```

In addition to the integrity of the target values across the 3 files, it's found that there are duplicated rows across the original dataset. Some noteable observations are:
- There are 95 duplicated rows considering only feature columns
- There are 46 and 65 duplicated rows considering feature columns and EC1~EC2 target columns from _ecfp file and _desc file respectively.
- This infers that in the original dataset with the same feature columns result in different target EC1 and EC2 outputs.
- This means that the explainibility of the features here are not strong enough to infer the target EC1 and EC2 values.
- This problem does not occur in the original train set, as proven in my EDA kernel.
